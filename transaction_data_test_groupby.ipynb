```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier as KNC
from sklearn.svm import SVC as SVC
from sklearn.tree import DecisionTreeClassifier as DTC
from sklearn.ensemble import GradientBoostingClassifier as GBC
from sklearn.ensemble import RandomForestClassifier as RFC
from sklearn.linear_model import SGDClassifier as SGDC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, mean_absolute_error, mean_squared_error
import matplotlib.pyplot as plt
from collections import defaultdict
%matplotlib inline
```

```python
transactions = pd.read_csv('./datasets/transactions/transactions.csv', nrows=500000)
```

```python
transactions.columns
```

```python
def dict_func(transactions):
    dic = defaultdict(lambda: 0)
    for row in transactions.itertuples():
        dic[row.customer_id] += row.amount
    return dic
```

```python
def agg_func(transactions):
    return transactions.groupby(by=["customer_id"]).aggregate({"amount": lambda x: np.sum(x)})
```

```python
%time dict_func(transactions)
%time agg_func(transactions)
```
